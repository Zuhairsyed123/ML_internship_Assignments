## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("/content/content_refresh_anonymized.csv")

total_pages = len(df)
declining_pages = (df["trend_direction"] == "down").sum()
decline_rate = (declining_pages / total_pages) * 100

print("Total rows:", total_pages)
print("Average impressions (90d):", round(df["impressions_90d"].mean(), 2))
print("Average sessions (90d):", round(df["sessions_90d"].mean(), 2))
print("Declining pages:", declining_pages)
print("Decline rate (%):", round(decline_rate, 2))
```

This project will provide decision-support recommendations based on observed search and engagement signals. By analyzing patterns in impressions, sessions, CTR, content age, and performance trends, it can help identify pages that may deserve review for refresh, expansion, protection, or monitoring.
